### Keywords: 
<pre>1. Keywords are reserved words in python that have a specific meaning and cannot be used as identifiers (variable names, function names, etc.).
2. There are 35 keywords in python (as of python 3.10) and they are: False, None, True, and, as, assert, break, class, continue, def, del, elif, else, except, finally, for, from, global, if, import, in, is, lambda, nonlocal, not, or, pass, raise, return, try, while, with, yield.
boolean = True
none_value = None</pre>

### Identifiers: 
<pre>1. They are the names used for variables, functions, classes, modules, or other objects in pythons. 
2. They must start with a letter (a-z, A-Z) or an underscore (_) followed by letters, digits (0-9), or underscores.
3. They cannot be a keyword.
4. They are case-sensitive (e.g., myVariable and myvariable are different identifiers).
5. We cannot use '-' in identifiers.</pre>
