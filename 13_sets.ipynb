```python
# initializing empty set 
set_ = set() # the set() is a inbuilt function 
print(type(set()))

set_var = {1, 2, 3, 4, 3, 4}
print(set_var)
```

## Inbuilt function of set. 

### 1. Add 

```python
set_var = {"Avengers", "Ironman", "Hulk"}
print(set_var)
set_var.add("Thor")
print(set_var)
```

## 2. Difference & Difference Update

```python
# Difference removes the common element from both the sets.
set1 = {1,2,3,4,5}
set2 = {1,2,3,4,5,6}

print(set1)
print(set2.difference_update(set1))
print(set2.difference(set1))
```

## 3. Intersection & Intersection Update

```python
# Intersection gives the common element between both set
set1 = {1,2,3,4,5}
set2 = {1,2,3,4,5,6}
print(type(set1))
print(type(set2))
print(set1.intersection(set2))
print(set1.intersection_update(set2))
```

## 4. Updates on Set

```python
print("Before updating set 1")
print(set1)
print("\n")
print("After updating set1")
set1.update((6,7,8,9,10))
print(set1)
```
