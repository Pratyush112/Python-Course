# Builtin Functions 
- There are total <strong>15</strong> different builtin functions.
1. print() -> Used to print data in the console.
2. input() -> Used to take user input 
3. type() -> Used to check the type of the data
4. int (), float(), string(), list(), tuple(), etc. -> These are the data types used to declare a variale and its data type in user input function.
5. abs() -> Absolute value or modulus of a number. Ex: abs(-5.6) = 5.6
6. pow() -> Power function -> pow(2, 3) = 2^3= 8
7. min()/max() -> Used to find the minimum or maximum in list, tuple, or sets
8. round() -> Round off function. Ex: a = 22/7 => round(a, 2) = 3.14 [a = value, 2 = no. of places we want the decimal to be round off.]
9. divmod -> Returns the tuples, that contains integer division value(//) and modulus of a value(%). Ex: divmod(5, 2) = (2, 1) -> 2 is the integer division (//) and 1 is the modulus of the value (%)
10. bin/oct/hex -> Returns the binary, octal and hexadecimal of a number. 
11. id -> returns the address of the variable inside memory
12. ord -> Return the ascii value of the character.
13. len -> Return the length of any iterable string, list, tuple, dictionary, sets etc
14. sum -> Returns the sum of iterables containing numerical values else it will show type error. 
15. help -> Returns the documentation of any builtin function.

```python
a = 1
list_1 = [1,2,3,4]
print("Type of a: ", type(a))
print("Absolute Value: ", abs(-4))
print("Power Function: ", pow(2, 3))
print("Minimum value in list_1: ", min(list_1))
print("Maximum value in list_1: ", max(list_1))
print("Round off value of 22/7: ", round(22/7, 4))
print("Divmod value: ", divmod(5, 2))
print("Binary Value of 56: ", bin(56), "\nOctal value of 56: ", oct(56), "\nHexadecimal value of 56", hex(56))
print("Address of the value in the memory: ", id(a))
print("Ascii value of 'b' is: ", ord('b'))
print("Length of the list_1 is: ", len(list_1))
print("Sum of the numbers in list_1: ", sum(list_1))
print("\nDocumentation of print() function we can get using help() function: \n")
help(print)
```
