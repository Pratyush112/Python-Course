# Loops 
- Loops in python are the control flow statement that allows to repeat a block of code multiple times. This is important for automating repetative tasks

### Types of Loops in Python are:
<strong>Do While:</strong>
- Repeats a block of code as long as a specific condition is true.
- The loop will execute at least once, even if the condition is initially false.
- Useful when number of iteration is unknown and depends on dynamic conditions.
- Initialise the loop before while loop condition.

<strong>Syntax:</strong>
<pre>
i = initialize value
while condition:
    # code block
</pre>

## One important concept is "Range Function" in python before starting For Loops we must know this:
- It is mostly used with For loop to iterate over a sequence of numbers.
- The range() function returns a sequence of number.
- It supports "Start", "Stop", "Step" parameters. -> range(start, stop, step)
- If we don't specify the start value, it automatically assumes the starting point to be 0
- If we don't specify the step value, it automatically assumes the step size to be 1.
- If we don't specify the stop value, it automatically generates infinite number of values.
- <strong>Example 1:</strong> range(1, 5, 2) it generates output as - 1, 3 does not includes the stop parameter.
- <strong>Example 2:</strong> range(1, 5) it generates output as - 1, 2, 3, 4 does not includes the stop parameter.
- <strong>Example 3:</strong> range(5) it generates output as - 0, 1, 2, 3, 4 since there is only one parameter, it assumes as it the stop value and starts from 0 and the step value to be 1.

<strong>For Loop: </strong>
- Used to iterate over a sequence (List, Tuple, string, sets, and range) or iterateable object.
- It automatically handles iteration and indexing, making it consise and readable.

<strong>Syntax:</strong>
<pre>
for i in range (start, stop, step):
    #code block
</pre>


### Example of While Loop -> Guessing Game:

- We will build a guessing game using While Loop
    - User have to guess the number 
    - If it guess the number higher than the actual value we will suggest to guess
    - If it guess the number lower than the actual value we will suggest to guess higher
    - Also we will count the number of iteration, how many attempt it took to guess the correct value. 

```python
import random
a = int(input("Enter the lower number: "))
b = int(input("Enter the upper number: "))
secret_number = random.randint(a, b)
guess_number = int(input("Guess a number: "))

count = 1
while guess_number != secret_number:
    if guess_number < secret_number:
        print("Guess Higher, Try again..")
    else:
        print("Guess Lower, Try again..")
    guess_number = int(input("Guess a number: "))
    count += 1

print(f"You took {count} attempts to guess the correct number")
```

### Example of For Loop in Python -> Calculate the price of items in cart with 10% discount

- We need list of items with price and the item name
- We will use for loop to iterate through each items in the cart

```python
cart_items = [
    {"name": "Laptop", "price": 10999},
    {"name": "Mouse", "price": 599},
    {"name": "Keyboard", "price": 1599},
    {"name": "Monitor", "price": 4999},
    {"name": "Headphones", "price": 2999}
]

total_price = 0

for items in cart_items:
    items["price"] *= 0.9
    total_price += items["price"]
    print(items)

#we could also rewrite the above code 
#for items in cart_items:
#    total_price += items["price"] * 0.9
#print(f"Total price after discount: Rs{total_price:.2f}")

print(f"Total price after discount: Rs{total_price:.2f}")
```

### Nested Loops
- We can also use nested loops to iterate through
- A "For loop" can contain "While loop"
- A "While Loop" can contain "For loop"
- Multiple "for loop" can be nested inside "for loops"
- Multiple "While loop" can be nested "While loops"  

## We will use patterns as an example for nested Loops in python 
<pre>
*
* *
* * *
* * * *
* * * * *
</pre>
- We are going to print this pattern using nested loops

```python
row = int(input("Enter the number of rows: "))

for i in range(1, row + 1):
    for j in range(1, i + 1):
        print("* ", end= " ")
    print()
        
```

### Break, Continue, Pass

- <Strong>Break: </strong> Terminates the loop at a specific condition
- <Strong>Continue:</strong> Skiping that particular block of code for particular iteration and continue with next iteration.
- <Strong>Pass:</strong> It is used to write empty blocks. i.e to skip a particular block of code but don't have the logic at this moment we can write pass.
