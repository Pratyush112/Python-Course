## Creating a Fraction DataType

### Methods we will give to this fraction class are 
1. add 
2. subtract
3. multiply
4. Division

````python
class Fraction:
    """ This is a fraction data type """

    def __init__(self, numerator, denominator):
        self.num = numerator
        self.den = denominator

    def __str__(self):
        return '{}/{}'.format(self.num, self.den)

    def __add__(self, other):
        new_num = self.num*other.den + self.den*other.num
        new_den = self.den*other.den
        # return '{}/{}'.format(new_num, new_den)
        return Fraction(new_num, new_den)

    def __sub__(self, other):
        new_num = self.num*other.den - self.den*other.num 
        new_den = self.den*other.den
        # return '{}/{}'.format(new_num, new_den)
        return Fraction(new_num, new_den)

    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den
        # return '{}/{}'.format(new_num, new_den)
        return Fraction(new_num, new_den)

    def __truediv__(self, other):
        new_num = self.num*other.den
        new_den = self.den*other.num
        # return '{}/{}'.format(new_num, new_den)
        return Fraction(new_num, new_den) # this is called factory function or factory methods, a design pattern, where a function or method act as a factory to instantiate and return a new object. 

    def convert_to_decimal(self):
        return self.num/self.den 
````

````python
obj1 = Fraction(1, 2)
obj2 = Fraction(3, 4)
print(obj1, '\n')
print(obj2, '\n')

fr1 = obj1 + obj2
fr2 = obj1 - obj2
fr3 = obj1 * obj2
fr4 = obj1 / obj2

print(fr1.convert_to_decimal(), '\n', fr2.convert_to_decimal(), '\n', fr3.convert_to_decimal(), '\n', fr4) 

print(type(obj1))
print(type(obj2))

````

### Create class for 2D geometry.
1. A user can create a view 2D co-ordinate 
2. A user can find out the distance between 2 co-ordinates
3. A user can find the distance of a co-ordinate form origin
4. A user can check if a point lies on a given line. 
5. A user can find the distance beetween a given 2D point and a given line. 
6. Two line segments intersect with each other or not.

````python
import collections
from math import sqrt

class coordinate:

    def __init__(self, x, y):
        self.x_coordinate = x
        self.y_coordinate = y

    def __str__(self):
        return '({},{})'.format(self.x_coordinate, self.y_coordinate)

    def distance_bw_Points(self, other): # Also known as Euclidean Distance
        return ((self.x_coordinate - other.x_coordinate)**2 + (self.y_coordinate - other.y_coordinate)**2)**0.5
        # return sqrt((self.x_coordinate - other.x_coordinate)**2 + (self.y_coordinate - other.y_coordinate)**2)

    def origin_distance(self): #In this origin co-ordinate is not mentioned explcitly.
        # return sqrt((self.x_coordinate)**2 + (self.y_coordinate)**2)
        return self.distance_bw_Points(coordinate(0,0))
````

````python
P0 = coordinate(0, 0)
P1 = coordinate(1, 1)
P2 = coordinate(3, 3)
print(P1) 
print(P2) 
print("\n", "The distance between Point P1(1, 1), and P2(2, 3)") 
print(P1.distance_bw_Points(P2))

print("\n", "The distance between Point P0(0, 0), and P1(1, 1)")  
print(P0.distance_bw_Points(P1))

print("\n", "The distance between Point P0(0, 0), and P2(2, 3)") 
print(P0.distance_bw_Points(P2))

print("\n", "The distance from origin P0(0, 0), and P1(1, 1)")  
print(P1.origin_distance())

print("\n", "The distance from origin P0(0, 0), and P2(2, 3)") 
print(P2.origin_distance())
````

## Reference Variable

```python
F = Fun()
```
- <b>Few things that is needed to be cleared here are</b>
1. When we call a class it creates a object in the memory.
2. We do not need any variable to make a object. 
3. But, we need a variable to store that objects memory reference id. Else that object will be lost in the memory. 
4. When we store the reference Id in a variable that variable starts pointing to that memory location. 
5. An object can have multiple reference variable. 
6. Assigning a new variable to the same object does not create a new object. Rather both the variable points to that same memory location. 

````python
class Fun:
    def __init__(self):
        self.park = 'Yes' 

Fun() # An object is created (2nd Point above)
print(id(Fun())) #here we are creating 2nd object, and printing its id, but this will be discarded as no variable holds it. 

F = Fun() # Creates another new object i.e 3rd -> Stores in Variable F. 
print("F: ",id(F)) # Printing the id of F. 
G = F # A new variable is getting assigned to the same object that is created 3rd time. Now it is pointing to the same memory location.
print("G: ",id(G)) #Here it will the same id as variable F. 

H = Fun() # Another new object will be created with different id. 
print("H: ",(id(H)))

print("F = G: ", F is G) #True
print("F = H: ",F is H ) #False
print(F.park)
print(G.park)

G.park = 'No'
print(F.park) 
print(G.park)
````

## Pass by Reference
- Passing objects reference to a function as argument. 
- Return object of a class from a function.

<div style="border:2px solid #af934cff; border-radius:6px; padding:20px; width: 90%; margin-top:10px;">
<b>NOTE:</b> <br> 
1. Passing the reference id of the object to the function, also we have checked that single object can have multiple pointers. (Shown in greet() function) <br>
2. Functions can return object of a class. (Shown in the greet_1() function) <br>
3. Objects are Mutable. (Shown in greet_2() function)
</div>

````python
class Person:
    def __init__(self, name, gender): #this is written inside the class hence it is a method. Generally known as constructor. 
        self.name = name
        self.gender = gender

#this is written outside the class, hence it is a function
def greet(person): #We have given objects address/reference as the parameter of the function. 
    print(f"Id of person: {id(person)}")
    print(f"Hello, I am {person.name} and I am {person.gender} ")

    print("\n")
    print("-> Here we will check, if we change one of the attribute inside function will it be changed outside the function or not")
    person.name = 'Ankur' 
    print("New name: ", person.name) # We have changed the value of the object through reference. 

# name = input("Enter your name: ")
# gender = input("Enter your gender: Male or Female")
p=Person('Pratyush', 'male')
print("Id of p and person(inside the function greet) are same.")
print("Id of p:", id(p))
greet(p)

print("-> This is for checking if the pointer 'p' is same as the pointer 'person'. Inside the function we have changed the value of the name attribute.") 
print("This will also print new name: ",p.name) # So the object reference outside the function will also show the updated value.

print("\n")

#We will return object from the function
def greet_1():
    print("Created a object inside the function")
    p1 = Person('Ankit', 'Male')
    return p1 

x = greet_1()
print("Returning object from function greet_1:", x.name)
print("Returning object from function greet_1:", x.gender)
# print(greet_1().name)
# print(greet_1().gender)

print("\n")

#this is the code for checking whether the objects are immuteable or not.
def greet_2(person1):
    person1.name = 'Ankur' 
    print("Id of Person1:",id(person1))
    return person1

print("This is where we check the mutability of the object.")
p1 = Person('Pratyush', 'male')
print("Prints the ID of p1 before sending the arguments to the function greet_2:", id(p1))
p2 = greet_2(p1)
print("Prints the id of p2:",id(p2))

````
