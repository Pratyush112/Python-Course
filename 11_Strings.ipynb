# Accessing Substring from a string

## Indexing

1. <b>Indexing:</b> It means that, we will iterate through the words and print the desired string. <br>
    <b>Example: </b> <br>
    a = "hello" <br>
    print(a[1]) -> here 1 refers to the letter that is present on the 1st position of the word 'hello'
    <br> 
    '*this is a +ve indexing (left -> right)*' 
    <br>
    <b>Output: </b> e <br>

    print(a[-1]) -> *This is called -ve indexing (right -> left)* <br>
    <b>Output: </b> o

2. <b>Slicing: </b> It will slice out the given part and print the rest of the part and that is how we can access the substring from a string
    <b>Example: </b>
    a = "hello world" 


```python
a = 'hello'
print(a[4])
print(a[-4])
```

## Slicing

<p>Syntax: 'string[start:end]'</p>
Some of the variations of Slicing are <br/>
<b>Important NOTE</b> left side of the colon is the starting point that is, from where the slicing will start and the right side of the colon is the ending point that is, the slicing will continue till that point. 

```python
a = "Hello World"
print(a[:]) #  Returns the complete string (no start or end specified)
print(a[0:5]) # Returns characters from index 0 up to (but not including) index 5
print(a[4:11]) #Returns characters from index 4 up to index 10 (the space at index 5, then "World")
print(a[:5]) # Returns from the beginning up to index 4 (same as a[0:5])
print(a[2:11]) # Returns characters from index 2 to the end of the string 
```

## Steps in Slicing

<b>Syntax: </b>string[start:end:step] <br/>
Definition: The step determines how many characters to jump at each iteration. Here are examples:

```python
a = "Hello World"

print(a[::2])      # HloWrd         (every 2nd character, entire string)
print(a[0:5:2])    # Hlo            (index 0 to 4, every 2nd character)
print(a[::1])      # Hello World    (every 1st character - same as a[:])
```

## Reverse of a string using Slicing

<b>Definition</b>: To reverse a string in Python, you can use string slicing with a step of -1. This creates a new string with the characters in reverse order.
<b>Syntax:</b> string[::-1] <br/><br/>
<b>NOTE:</b> While counting from reverse we start from 1 and not 0. <br/><br/>
<b>Example:</b> "Hello World" - if we count from reverse then - d = 1st index, l = 2nd index, r = 3rd index, 0 = 4th index, W = 5th index, o = 6th index, l = 7th index, l = 8th index, e = 9th index, H = 10th index

```python
a = "Hello World"
print(a[::-1])     # dlroW olleH    (reverse string, step -1)
print(a[10:0:-1])  # dlroW          (reverse from index 10 to 1)
# We can also reverse a substring
print(a[5:0:-1])  # Output: olleH (reverses from index 5 to 1)
```

# Edit & Delete Strings 

### - Editing

<b>NOTE: </b> Strings are immuteable and hence, editing of strings directly on the original string are not applicable in python

### - Deleteing

<b>NOTE: </b> This operation is also not allowed in directly on the original string as the string is immuteable

### What we can do is

1. Re-assign in the values of the string by creating a new string, the new string will be based on the original one.
2. Completely delete that string in place of partial deletion.

```python
print("Editing of the string using 3 method, by making a new string\n1. .replace()\n2. Slicing\n3. Converting to a list and Back \n")
print("Below is the Modified or edited string: \n")
text = 'hello world' #This is the original string

print("Editing string using .replace() method\n")
text1 = text.replace('hello','hey') # This is the new string based on the original one, and we've edited the old string
print("output 1: ", text1, "\n")

print("Editing string using Slicing method\n")
text2 = text[:5] + ' Python' #using slicing and concatenation method
print("Output 2: ",text2, "\n")

print("Editing string by converting to a list and back\n")
text3 = list(text)
text3[6] = "J" #Change 'w'
text3[7] = "A" #change 'o'
text3[8] = "V" #change 'r'
text3[9] = "A" #change 'l'
text3[10] = "!" #change 'd'

new_text = "".join(text3)
print("Output 3: ", new_text, "\n")
```

### Relational Opearator
<b>1. ==</b> <br/>
<pre> a = "Hello" == "World"
<b>Output</b>: False
</pre>
<b>2. !=</b> <br/>
<pre> a = "Hello" != "World"
<b>Output</b>: True

<b>3. > </b> <br/>
<pre> a = "Hello" > "World"
<b>Output</b>: False
</pre>
<b>4. <> </b> <br/>
<pre> a = "Hello" < "World"
<b>Output</b>: True
</pre>

```python
print("String Operations: \n")
print("1. Arithmetic Operations: \n")
print("a. String Addition (Concatenation): \n")
print("Output of String Addition:", "Hello"+"World", "\n") #This will concatenate the two strings together
print("b. String Multiplication (Repetition): \n")
print("Output of String Multiplication:", "Hello "*3, "\n") #This will repeat the string "Hello " three times
print("2. Comparison/Relational Operations: \n")
print("a. Equality (==): \n")
print("Output of Equality (Hello == World):", "Hello" == "World", "\n") #This will return False because the strings are different
print("b. Inequality (!=): \n")
print("Output of Inequality (Hello != World):", "Hello" != "World", "\n") #This will return True because the strings are different
print("c. Greater than (>): \n")
print("Output of Greater than (Hello > World):", "Hello" > "World","They follow the lexicographical order", "\n") #This will return False because "Hello" is not greater than "World" in lexicographical order
print("d. Less than (<): \n")
print("Output of Less than (Hello < World):", "Hello" < "World","They follow the lexicographical order", "\n") #This will return True because "Hello" is less than "World" in lexicographical order
print("e. Greater than or equal to (>=): \n")
print("Output of Greater than or equal to (Hello >= World):", "Hello" >= "World", "They follow the lexicographical order", "\n") #This will return False because "Hello" is not greater than or equal to "World" in lexicographical order
print("f. Less than or equal to (<=): \n")
print("Output of Less than or equal to (Hello <= World):", "Hello" <= "World", "They follow the lexicographical order", "\n") #This will return True because "Hello" is less than or equal to "World" in lexicographical order
```

```python
print("Hi my name is Pratyush".endswith("Pratyush")) # True : because it is the last word
print("Hi my name is Pratyush".endswith("P")) # False: Because it is not the last letter
print("Hi my name is Pratyush".endswith("h")) # True: Because it is not the last letter
print("Hi my name is Pratyush".startswith("H")) # True: Because It start with that letter
print("Hi my name is Pratyush".startswith("Hi")) # true: Because it start with that word
print("Hi my name is Pratyush".startswith("Pratyush")) # Flase: Because it does not starts with that word.
```

<pre>5. isalnum/isalpha/isdecimal/isdigit/isidentifier
alnum - alphanumeric
aplha - alphabet
decimal - (1 - 10)
digit - it contains all the superscript and subscript, but not fractions and Roman numbers
numeric - It contains all the basic numbers, subscript/superscript numbers, fraction numbers, roman numbers.
identifier - Recognises uderscore, dashes etc
They also return true or false</pre>

```python
print("isalnum()")
print('flat20'.isalnum()) 
print('flat20&'.isalnum()) 
print('\n')
print("isalpha()") 
print('flat20'.isalpha())
print('flat'.isalpha())
print('\n')
print("isdigit()/isdecimal()")
print('20'.isdigit())
print('20b'.isdigit())
print('\n')
print("isidentifier()")
print("hello world".isidentifier())
print("Hello_world".isidentifier())

print("\n")
print("Difference between isdigit() and isdecimal()")
s = "²"  # Unicode superscript 2 (\u00B2)
print(s.isdigit())
print(s.isdecimal())
```
