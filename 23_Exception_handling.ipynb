1. **Before the program runs (Compilation)** (Syntax Errors)
2. **While the program is running (Execution)** (Exceptions)

---

Think of Python like a spoken language. If you spell a word wrong or forget a punctuation mark, Python simply won't understand you.

- **Who catches it:** The Python interpreter/Compiler catches it *immediately*, before the code even tries to run.
- **How to fix it:** Read the error message, find the typo, and correct the code. It's an easy fix!

# `try-Except` Block:
## Exceptions (Runtime Errors)
- **When do they happen?** During the execution of the program (at runtime).
- **Why do they happen?** The grammar of your code is perfectly fine, but something unexpected occurred (like trying to open a file that doesn't exist).
- **Who raises them?** The Python runtime environment.
- **How to fix them?** You have to tackle them "on the fly" by writing logic that anticipates and handles these unexpected situations.

### Common Examples:
1. **Memory Overflow** (Running out of RAM)
2. **ZeroDivisionError** (Trying to divide a number by 0)
3. **Database Errors** (Losing connection to a database)

---
### Why is Exception Handling Important?
1. **Better User Experience (UX):** Python's default red error messages look scary and confusing to non-technical users. Handling exceptions allows us to show clean, friendly error messages instead.
2. **Security:** Default error tracebacks often reveal sensitive information about your code's structure. Hackers can use this information to find vulnerabilities and take control of your system.

---
### How to Handle Exceptions: The `try-except` Block

To protect our code from crashing, we put the "risky" code inside a `try` block. If an error happens, Python immediately jumps to the `except` block instead of crashing!

```python 
try:
    with open('sample.txt', 'r') as f:
        print(f.read()) 
except:
    print('File not found! Please check the filename.')
```
---
### Catching Specific Errors
Different lines of code can cause different types of errors. Instead of using one generic except block, it is best practice to handle specific errors individually. At the very end, we can add a "catch-all" generic exception just in case we missed something.


> **Pro Tip:** **Understanding the Error** 
> If you want to see exactly what went wrong without crashing the program, you can catch the error as a variable using `except Exception as error:`
> 
> ```python
> try:
>     print(s)
> except Exception as error:
>     print(error.with_traceback)
> ```
> *Output: <built-in method with_traceback of NameError object at 0x0000023A082097E0>*
> 
> ```python
> try:
>     print(s)
> except Exception as error:
>     print(f"What happened: {error}")
>     print(f"Type of error: {type(error).__name__}")
> ```
> *Output: What happened: name 's' is not defined Type of error: NameError*
>
> `print(error.with_traceback)` this line shows the type of error, like, NameError, or ValueError, etc. While `except Exception as error: print(error)` this line show what error has occured.

> **Important Rule: The Catch-All Goes Last!** `except Exception as error:` should always be written last. If placed at the top, this line will immediately take over whenever any error occurs, meaning the specific errors you defined below it will never be checked or shown. 

> **A small Analogy for the above Rule:** Think of it like a giant net that catches all errors. If you put the giant net first, it catches the error immediately, and your specific, smaller nets (like `NameError` or `FileNotFoundError`)  underneath it will never get a chance to work!

### else - Finally: 
**Else:** The control comes to else, when there is no problem or no error in the try block.
**Finally:** 
- The Finally block always runs, no matter what, If there is an error in the try block, then exception will execute and then the Finally block will run.
- If there is no error in the try block, then else block will execute and then finally block will execute. 
 

````python
# Executing the the exception and Finally block. 
try: 
    file = open("22_file_handling_output/demo8.txt", "r")
except FileNotFoundError:
    print("No such file is present, please create one first")
else:
    print(file.read())
finally:
    print("Finally block executed")
````

````python
# Executing the the else and Finally block. 
try: 
    file = open("22_file_handling_output/demo3.txt", "r")
except FileNotFoundError:
    print("No such file is present, please create one first")
else:
    print("No error in try block, else executed ---> ", file.read())
finally:
    print("Finally block executed")
````

### Raising Custom Errors (The `raise` Keyword)

Normally, Python crashes when *it* finds a mistake (like dividing by zero). But what if *you* want to trigger an error on purpose when someone breaks a rule you created? 

That's exactly what the `raise` keyword does! It allows you to **throw your own custom errors**.

---

#### The Scenario: Why do we need `raise`?
Imagine you are building a banking app, and you have a function to withdraw money.
Python doesn't know anything about banking rules. If a user tries to withdraw a **negative amount** (like `-50` dollars), Python's math engine thinks this is perfectly fine. 

But *you* know it's a rule violation! You can use `raise` to stop the program and shout, *"Hey! That's not allowed!"*

#### Behind the Scenes: What is actually happening?
When you write code like `raise Exception("Amount cannot be negative!")`, here is what happens step-by-step:

1. **Creating an Object:** You are actually using Object-Oriented Programming! `Exception("...")` creates a brand new **Error Object** in the computer's memory. The string you passed is saved inside this object as the error message.
2. **Throwing the Alarm:** The `raise` keyword grabs that Error Object and forcefully throws it into the running program like a fire alarm.
3. **The Search for a Net:** Once the alarm is raised, Python immediately drops whatever it was doing and starts frantically looking for a `try-except` block to catch the error. 
4. **Crash (if not caught):** If Python cannot find an `except` block to catch your custom error, the program crashes, and your custom message is printed in red text on the screen.

#### Demo: Enforcing Rules with `raise`

```python
class BankAccount:
    def withdraw(self, amount):
        if amount < 0:
            # We are intentionally causing an error here!
            raise ValueError("Amount cannot be negative!")
        else:
            print(f"Successfully withdrew ${amount}")

# Let's test our rule!
account = BankAccount()

try:
    account.withdraw(-50) # This breaks our rule!
except ValueError as error:
    print(f"Transaction Failed: {error}")

```

>**Pro Tip:** Notice how we used raise ValueError(...) instead of raise Exception(...). While Exception works, it is best practice to raise a specific built-in error (like ValueError or TypeError) so other programmers know exactly what kind of rule was broken!

````python
class Bank:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount < 0:
            raise Exception("Amount cannot be negetive")
        if self.balance < amount:
            raise Exception("Insufficient Balance")
        self.balance -= amount
    
obj = Bank(20000)
try:
    obj.withdraw(25000) # Throws error of insufficient balance
except Exception as e:
    print(e)
else:
    print(obj.balance)
````

### Creating Custom Excetions
Sometimes, built-in errors like `ValueError` or `TypeError` just don't fit the specific rule you are trying to enforce in your program. 

Luckily, because Python is Object-Oriented, you can create **your very own Exception**!

- **How?** You simply create a new Class and make it inherit from Python's built-in `Exception` class.
- **Why?** It makes your code incredibly easy to read. If another programmer sees `raise InsufficientFundsError`, they know *exactly* what went wrong!

#### The Simplest Way to Create a Custom Exception
Because the parent `Exception` class already knows exactly how to handle and store error messages, all you need to do is inherit from it and use the `pass` keyword!

```python
class CustomException(Exception):
    def __init__(self, message):
        print(message)
class Bank:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount < 0:
            raise CustomException("Amount cannot be negetive")
        if self.balance < amount:
            raise CustomException("Insufficient Balance")
        self.balance -= amount
    
obj = Bank(20000)
try:
    obj.withdraw(-5000)
except CustomException as e:
    pass
else:
    print(obj.balance)
```

*Output: Amount cannot be negetive*
