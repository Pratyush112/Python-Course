## Encapsulation

- It is a fundamental concept in OOP, it bundles attributes and methods that operate on the attributes into a single class. This helps to protect the internal state of an object from unintended interference and misuse.
- Here are some key points about encapsulation:
    - **Data Hiding**: We can hide the internal state of an object and only expose what is necessary through public methods. The hidden data can be accessed and modified only through getter and setter methods.
    - **Access Modifiers**: In many programming languages, access modifiers (like __public, private, and protected__) are used to control the visibility of class members. This allows developers to restrict access to certain parts of a class.
    - **Improved Maintainability**: By encapsulating the data and methods, we can change the internal implementation of a class without affecting the external code that uses it. This makes the code easier to maintain and evolve over time.
- **Representation** of public, private, and protected members in a class:
    - **Public members**: Accessible from anywhere in the program. (self.attribute)
    - __Private members__: Accessible only within the class itself. (self.__attribute)
    - __Protected members__: Accessible within the class and its subclasses. (self._attribute)


```python
class Example:
    def __init__(self):
        self.public_attribute = "I am public"
        self.__private_attribute = "I am private"
        self._protected_attribute = "I am protected"
```
**In Memory Representation**: 
1. The `public_attribute` is stored same as the variable name.
2. The `_protected_attribute` is also stored with the same name but is conventionally treated as protected.
3. The `__private_attribute` is name-mangled to `_Example__private_attribute` -> This is the reason, why it is not directly accessible from outside the class.

### Collection of object

- We can collect objects in multiple ways:
1. list 
2. Dictionary 
3. set
4. tuple

- We can do all the operation in that object, which we perform using the list, dictionary, set, tuple.

````python
class Person:
    def __init__(self, name, gender, age):
        self.name = name 
        self.gender = gender 
        self.age = age 

p1 = Person('Pratyush', 'male', '22')
p2 = Person('Ankur', 'male', '23')
p3 = Person('Abhinav', 'male', '24')
p4 = Person('Abhinash', 'male', '25')
p5 = Person('Kunal', 'male', '26')

d = [p1, p2]
print([p.name for p in d])
print(type(d))

f = {p1, p2, p3, p4}
print(f)
print(type(f))

g = (p1, p2, p3, p4)
print(g)
print(type(g))
````

### Doing all the __list__ operation in the person object

1. append()
2. copy()
3. clear() 
4. count()
5. extend()
6. index()
7. insert()
8. pop()
9. remove()
10. reverse()
11. sort()

````python
#1. append(): adds a single element to the end of the list
d.append(p3) 
print('1. append(p3):', [p.name for p in d])

#2. copy(): Returns a shallow copy of the list
h = d.copy()
print('2. copy(): ', [p.name for p in h])

#3. count(): Returns the number of times an element appears
print('3. count(): ', d.count(p1))

#4. extend(): Adds all elements of another list to the end
d.extend([p4])
print('4. extend():', [p.name for p in d])

#5. index(): Returns the index of the first occurrence
print('5. Index', d.index(p3))

#6. insert(index, value): Inserts an element at a specific index
d.insert(1, p3) # 1st one is index and 2nd one is value
print('6. insert(1, p3)', [p.name for p in d])

#7. pop(): Removes and returns the element at given index (default: last)
popped = d.pop()
print('7. pop():', popped.name, '|List now:', [p.name for p in d])

#8. remove(): Removes the first occurance of an element
d.remove(p3) #Removes the first p3
print('8. remove(p3):', [p.name for p in d])

#9. Reverse(): Reverses the list in-place
d.reverse()
print('9. reverse():', [p.name for p in d])

#10. sort(): Sorts the list in-place (needs key since Person objects aren't directly comparable)
d.sort(key=lambda person:person.name)
print('10. sort(by name):', [p.name for p in d])

#11. clear(): Removes all elements from the list
d.clear()
print('11. clear():', d)
````

### Doing all the __dictionary operation__ in the Person object. 

1. keys()
2. values()
3. items()
4. get()
5. update()
6. setdefault()
7. pop()
8. popitem()
9. Copy()
10. fromkeys()
11. Clear()

````python
e = {"p1": p1, "p2": p2, "p3": p3, "p4":p4}

#1. Keys(): Returns all the keys of the dictionary
print('1. keys():', list(e.keys()))

#2. values(): Returns the value of the dictionary
print('2. values():', [p.name for p in e.values()])

#3. items(): Returns all the values of the dictionary
print('3. items():', [(k, v.name) for k,v in e.items()])

#4. get(): Returns the value for a key(returns None if Key dosen't exist, no error)
print("4. get('p1'):", e.get("p1").name)
print("get('p99'):", e.get("p99")) #Returns None instead of creashing

#5. update(): Updates the dictionary with new key-value pairs 
e.update({"p5": p5})
print("5. update({'p5':p5}):", [p.name for p in e.values()])

#6. setdefault(): Returns value if key exists, else inserts key with default value
result = e.setdefault("p1", p5) #p1 exists so returns the existing value
print("6. setdefalut('p1'):", result.name, "(existing key, no change)")
result2 = e.setdefault("p6", p3) #p6 dosen't exist, so it inserts p3 as value.
print("setdefault('p6'):", result2.name, "(new key, inserted)")
print("dict now:", list(e.keys())) 

#7. pop(): Removes and returns the value for a given key
popped = e.pop("p6")
print("7. pop('p6'):", popped.name, "|keys now:", list(e.keys()))

#8. popitem(): Removes and returns the last inserted key-value pair
last_key, last_value = e.popitem()
print("8. popitem():", last_key, "->", last_value.name, "|Keys now:", list(e.keys()))

#9. copy(): Returns a shallow copy of the dictionary
e_copy = e.copy()
print("9. copy():", list(e_copy.keys()))

#10. fromkeys() - Creates a new dictionary with given keys and a default value
new_dict = dict.fromkeys(["a", "b", "c"], "defalut_value")
print("10. fromkeys():", new_dict)

#11. clear(): Removes all items from the dictionary
e.clear()
print("11. clear():",e)
````

## Static Variables  

- These variables are same for all the objects. 
- Written inside the class, but outside all the methods.
- Syntax: <classname.variable_name>

````python
class Counter:
    counter = 1
    def __init__(self):
        self.count = Counter.counter
        Counter.counter += 1

c1 = Counter()
c2 = Counter()
c3 = Counter()

#The count will be incremented by 1, and this is called static variable, since the counter variable is same for all the objects.
print(c1.count)
print(c2.count)
print(c3.count) 
print(Counter.counter)
````

## Instance variable
- These are unique for all the objects
- Written inside the class, inside methods (constructor)
- Syntax: <self.variable_name>

````python
class Counter:
    
    def __init__(self):
        self.counter = 0
        self.counter += 1

c1 = Counter()
c2 = Counter()
c3 = Counter()

#Here the counter variable is unique for all the objects, this is called instance varibale. 
print(c1.counter)
print(c2.counter)
print(c3.counter)
````

````python
class Counter:
    __count = 1
    def __init__(self):
        self.counter = Counter.__count 
        Counter.__count += 1
    
    # utility functions, which do not need any objects as parameter
    @staticmethod 
    def get_count():
        return Counter.__count


c1 = Counter()
c2 = Counter()
c3 = Counter()

Counter.get_count()
````
