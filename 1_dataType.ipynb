<h2>Python supports 3 categrories of data types:</h2>
<ol>
<li> Basic Data Types: int, float, boolean, complex, string</li>
<li>Container Types: List, Tuples, Sets, and Dictionary</li>
<li>User Defined Types: Class and Object</li>
</ol>

<h4><strong>Basic Data Types:</strong></h4> 
    <ol>
    <li> <strong>int:</strong> Whole number without a decimal point. It can be positive, negative, or zero. (range: -1e308 to 1e308)</li>
    <li><strong>float:</strong> Number with a decimal point. It can also be in scientific notation. (range: -1.7e308 to 1.7e308)</li>
    <li><strong>boolean:</strong> Represents one of two values: True or False.</li>
    <li><strong>complex:</strong> Represents a complex number with a real and imaginary part. (e.g., 3 + 4j)</li>
    <li><strong>string:</strong> A sequence of characters enclosed in single, double, or triple quotes. (e.g., 'Hello', "World", '''Python''')</li>
    </ol>

```python
print("Int: type(42) -> ",type(42)) # Output: <class 'int'>
print("Float: type(3.14) -> ",type(3.14)) # Output: <class 'float'>
print("Boolean: type(True) -> ",type(True))  # Output: <class 'bool'>
print("Boolean: type(False) -> ",type(False))  # Output: <class 'bool'>
print("Complex: type(3 + 4j) -> ",type(3 + 4j))  # Output: <class 'complex'>
print("String: type('Hello, World!') -> ",type("Hello, World!"))  # Output: <class 'str'>
```

<h4><strong>Container Types:</strong></h4>
<ol>
<li>List: An ordered, mutable collection of items. It can contain elements of different data types. (e.g., [1, 'Hello', 3.14])</li>

<li>Tuple: An ordered, immutable collection of items. It can also contain elements of different data types. (e.g., (1, 'Hello', 3.14))</li>

<li>Set: An unordered collection of unique items. It is mutable and does not allow duplicate values. (e.g., {1, 2, 3})</li>

<li>Dictionary: An unordered collection of key-value pairs. It is mutable and allows for fast lookups based on keys. (e.g., {'name': 'Alice', 'age': 30})</li>  
</ol>

```python
print("list: ([1, 'Hello', 3.14]) -> ",type([1, 'Hello', 3.14]))  # Output: <class 'list'>

print("tuple: ((1, 'Hello', 3.14)) -> ",type((1, 'Hello', 3.14)))  # Output: <class 'tuple'>

print("set: ({1, 2, 3}) -> ",type({1, 2, 3}))  # Output: <class 'set'>

print("dict: ({'name': 'Alice', 'age': 30}) -> ",type({'name': 'Alice', 'age': 30}))  # Output: <class 'dict'>
```

<h4><strong>User Defined Types:</strong></h4>
    <ol>
        <li><strong>Class:</strong> A blueprint for creating objects. It defines a set of attributes and methods that the objects created from the class will have. (e.g., class Person: ...)</li>
        <li><strong>Object:</strong> An instance of a class. It is created using the class and can have its own unique attributes and behaviors. (e.g., person1 = Person())</li>
    </ol>

```python
class Person: 
    def __init__(self, name, age):
        self.name = name
        self.age = age
print(type(Person))  # Output: <class 'type'>
```

```python
person1 = Person("Alice", 30)
print(type(person1))  # Output: <class '__main__.Person'>
```
