<h1><strong>If Else Statemen</strong></h1>
<p>If else statement in python is used for decision making. It executes a block of code if the condition is true, and executes the another block of code if the condition is false </p>
<strong>Syntax:</strong>
<pre>
    if condition:
        # Code to execute if the condition is true
    else:
        # Code to execute if the condition is false
</pre>

### Example 1: If Else Example below


```python
age = int(input("Enter your age: "))
if age>= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

<h1><strong>If Else elif Statement</strong></h1>
<p>If else statement: If else statement in python is used for decision making. It executes a block of code if the condition is true, and executes the another block of code if the condition is false </p>
<p>If Else Elif Statement: <strong>Elif</strong> is the short form of "else if". It allows to check multiple condition</p>
<strong>Syntax:</strong>
<pre>
    if condition: 
        #code block
    elif condition: 
        #code block
    else:
        #code block 
</pre>

### Example 2: If Else Elif Example below

```python
age = int(input("Enter your age: "))
if age <= 5:
    print("You are a minor and will not be charge for the ticket.")
elif age > 5 and age < 18: 
    print("You will be charged the Half ticket price")
elif age >= 18 and age < 60:
    print("You will be charged the full ticket price")
else: 
    print("You are a senior citizen and will be granted special seat")
```

<h1><strong>Nested if else Statement</strong></h1>
<p>In this a block of "If else elif" is placed inside another "If else elif" block, which allows to make complex decision-making logic, where a condition is checked only if a prior condition is true.</p>
<strong>Syntax:</strong>
<pre>
    If Condition:
        If condiiton:
            # code block
        elif condition:
            # code block
        else:
            #code block
    else:
       # code block
</pre>

### Example of Nested if else Statement

```python
score = int(input("Enter your score: "))
if score >= 40:
    if score >= 90:
        print("You passed with distinction")
    elif score >= 70:
        print("You passed with A grade")
    else: 
        print("You passed with B grade")
elif score >= 20:
    print("You passed with C the exam")
else: 
    print("You failed the exam")
```

# Email & Password Validation system Example

- This program will check if an email address is valid or not
- A valid email address should contain the "@" symbol and a "." symbol after the "@" symbol. 

```python
# This program will check if an email address is valid or not
# A valid email address should contain the "@" symbol and a "." symbol after the "@" symbol. 

email_input = input("Enter your email address: ")

if "@" in email_input:
    if "." in email_input and ".com" in email_input:
        print("The email address is valid.")
    else:
        print("Email Address is not valid")
else:
    print("The email address is invalid.")
    
```
