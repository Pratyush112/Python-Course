## List indexing
- Each element in the list has a specific index. We can accss elements in a list via indexing. 
- The fisrt item in the list has an index of 0, second item has index of 1, and so on. 
- -ve indexing is when we refer to items from the end -1 refers to the last item, -2 refers to the second last item and so on. 

|      | apple | banana | mango |
|------|-------|--------|-------|
| index| 0     | 1      | 2     |
|------|-------|--------|-------|
|-ve index| -3 |  -2    | -1    |

```python
myList01 = ["apples", "bananas", "mango"]
print(myList01[0])
print(myList01[1])
print(myList01[2])
print("\n")

print(myList01[-1])
print(myList01[-2])
print(myList01[-3])


```

## List Sclicing 
- To access a range of items in a list, we use the slice operator.
- During slicing, when we mention the stop index, it does not include that element in the list. 

### Syntax 
`
list[start index : stop index]
`

`list[0:3]`

```python
myList = ["apples", "bananas", "mango", "cherry", "dragonfruite"]

#Accessing the first item in the list
print(myList[0])

#Accessing the first 3 elements
print(myList[0:3]) 

#Accessing the last 2 elements
print(myList[3:5]) 
```

## Adding elements

- We can add elements to a list using the `append`, `extend` and `insert` methods. Below is how we can apply the methods to the list. 

## Append
- We can only add a single element at the end of the list. 

## Extend 
- We can add multiple items at the end of the list.

## Insert 
- We can insert a single element at a specific position in a list using the insert method. 


### Syntax
`mylist.append("element_name")`
`mylist.extend(["Element_name", "Element_name"])`
`mylist.insert(2, "Guava")`

```python
myList = ["apples", "bananas", "mango"]
print("Print the existing list:",myList, "\n")
myList.append("dragonfruit")
print("add the dragonfruit at the end:",myList, "\n")

myList.extend(["dragonfruite", "Guava"])
print("Adds the list of fruite at the end:",myList, "\n")

myList.insert(2, "cherries")
print("Insert Cherries at 2nd position:",myList)
```

## Removing Element
- We can remove elements from a list using the `remove`, `pop`, and `delete` methods. 


## Remove
We can remove a single element as an arguement. 

### Syntax 
`mylist.remove("Element_name")`

## Pop 
- We can remove an element by referring to its index number. pop() return the removed element

### Syntax 
`mylist.pop(<index_number>)`

## Delete
- We can delete an element by referring to its index number. 

### Syntax
`del(myList[2])`


```python
myList = ["apple", "banana", 3, 4, 5]
print(myList, "\n")

myList.remove(4)
print("using Remove method:",myList, "\n")

removed_element=myList.pop(1)
print("using Pop element: ", myList)
print(removed_element)

del(myList[1])
print("using delete method: ", myList)
```
