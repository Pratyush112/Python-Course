# Abstraction 

- **Basic Meaning:** Hiding the internal implementation details and exposing only the necessary functionalities to the user.
- **Purpose:** Used to achieve data abstraction by defining a **common interface (contract)** for all of its subclasses.
- **Why use it?** It enforces **consistency** across large projects. If multiple developers are building different subclasses, abstraction guarantees they all use the exact same method names!

> **💡 Real-World Analogy:**  
> Think of a **TV Remote** or an **ATM**. You just press the "Power" or "Withdraw" button. You know *what* it does (the interface), but you don't need to know the complex circuitry inside or *how* it works (the implementation).
> 
> **Graphic Software Example:** A software has many shapes (Circle, Rectangle, Triangle). They all share a common action like `draw()`. We care about calling `draw()` on a shape, not how each specific mathematical shape is rendered behind the scenes.

---

## Structure of Abstraction 
- We need `ABC` (Abstract Base Class) and the `@abstractmethod` decorator imported from Python's built-in `abc` module. 
- We create an **abstract method** inside the abstract class. This method contains **no logic** (just `pass`). This forces any subclass to provide its own implementation, ensuring a consistent structure across all derived classes.
- **The Rule of Instantiation:** Since an abstract class contains incomplete (abstract) methods, **we cannot instantiate an abstract class directly** (doing so raises a `TypeError`). We can only instantiate subclasses that have implemented all the abstract methods.
- *(Note: An abstract class can also contain normal methods with logic that subclasses can inherit and use directly!)*

---

## Implementation of Abstraction 

**Step 1:** Import the `ABC` class and the `abstractmethod` decorator from the `abc` module.
```python
from abc import ABC, abstractmethod
```
2. Create an abstract class by passing `ABC` as its parent parameter.

__code__:  
```python
from abc import ABC, abstractmethod
class Animal(ABC):
```
3. Define a method inside the class, add the `@abstractmethod` decorator directly above it, and leave the body empty using `pass`.

__code__: 
```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def species(self):
        pass # Abstract method
```
4. Create a subclass that inherits from the abstract class and overrides the abstract method with actual implementation logic. Now, we can create an object of the subclass!

__code__: 
 ```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def species(self):
        pass  # Abstract method
# Subclass implementing the contract
class Dog(Animal):
    def species(self):  # Providing the required implementation logic
        return "Husky"
# We cannot do: animal = Animal()  <-- This would raise a TypeError!
# Instead, we instantiate the subclass:
dog = Dog()
print(dog.species())  # Output: Husky
```

````python
#importing Abstract Base Class from abc module
from abc import ABC, abstractmethod 

class Bank(ABC): #passing ABC as parameter
    @abstractmethod
    def pin(self):
        pass 

#Implemented without the abstracted method
class AppBank(Bank):
    def withdraw(self):
        print("Withdrawing money...")

#If we try to create the object, it will show a type error. 
bank = AppBank()
````

````python
#importing Abstract Base Class from abc module
from abc import ABC, abstractmethod 

class Bank(ABC): #passing ABC as parameter
    @abstractmethod
    def pin(self):
        pass 

#Implemented with the abstracted method, therefore no TypeError message. 
class AppBank(Bank):
    def pin(self):  #This is the asbtracted method from the parent class, this method helps instantiate the abstracted class. 
        print ("Enter you pin")

    def withdraw(self):
        print ("Withdrawing money...")

#If we try to create the object, it will show a type error. 
bank = AppBank()
bank.pin()
bank.withdraw()
````

````python
#lets see one example with property
from abc import ABC, abstractmethod

class Animal:
    @property
    @abstractmethod
    def species(self):
        pass # Abstract method

class Dog(Animal):
    @property
    def species(self): #The abstract method from Animal class with implementation logic.
        print("husky")

dog = Dog()
#since species is a property, we do not need '()' to call the species proper of Dog. 
dog.species
````
