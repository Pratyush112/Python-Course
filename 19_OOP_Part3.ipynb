### Aggregation
- Meaning -> something has a relationship -> One class own the other class
- Example: 
    1. (Customer) has a (Address). Means there is a relationship between a customer and address. 

#### Class Diagram

```mermaid
classDiagram

        +edit_profile()
    }
    class Address {
        -city
        +state
```

> **◆ = Aggregation**: Customer *owns* an Address object

````python
# Taking the example of customer and address 

class Customer:
    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender 
        self.address = address 

    def print_address(self):
        print(self.name, self.address.get_city(), self.address.state, self.address.pincode)

    def edit_profile(self, new_name, new_city, new_state, new_pincode):
        self.name = new_name
        self.address.edit_address(new_city, new_state, new_pincode)

class Address:
    def __init__(self, city, state, pincode):
        self.__city = city
        self.state = state
        self.pincode = pincode

    def get_city(self):
        return self.__city

    def edit_address(self, new_city, new_state, new_pincode):
        self.__city = new_city
        self.state = new_state
        self.pincode = new_pincode

    

address = Address('Tarakeswar', 'West Bengal', '111000')
c1 = Customer('Pratyush', 'Male', address)

c1.print_address()

c1.edit_profile('Adarsh', 'Kolkata', 'West Bengal', '111110')
c1.print_address()
````

## Constructor
- We can inherit constructor from parents to child
### Situations
1. When Child class does not have it's own constructor. Then through inheritence, we can call the parents class constructor. 
2. When child class have it's own constructor. Then by default it will call it's own constructor not the parents class constructor. 


````python
# here child class did not had it own constructor

class Phone:
    def __init__(self, price, brand, camera):
        print("inside Phone constructor")
        self.price = price 
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying Phone")

    # def get_buy(self):
    #     return self.__buy()

class SmartPhone(Phone):

    def buy(self):
        print("Buying smartphone")

s = SmartPhone(2000, 'realme', 12)
# s = Phone(2000, 'Realme', 12)
s.buy()
s.brand
# s.brand
````

````python
# Here child class has its own constructor, and it will call its own constructor 
# Can still access parents class methods not the attributes. 
class Phone:
    def __init__(self, price, brand, camera):
        print("inside Phone constructor")
        self.price = price 
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Parents Class: buying Phone")

class SmartPhone(Phone):
    def __init__(self, os, ram):
        print("Inside Smartphone constructor")
        self.os = os
        self.ram = ram

    def buy1(self):
        print("Child Class: Buying smartphone")

s = SmartPhone('Android', 12)
s.buy()
s.buy1()
````

## Method Overriding

- If Parents and child class have the same methods with same name, by default child class method will be called.

````python
#This is called method overriding
class Phone: 
    def __init__(self, price, brand, camera):
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("parent Class: Buying Phone")

class SmartPhone(Phone):
    def __init__(self, os, ram):
        self.os = os
        self.ram = ram

    def buy(self):
        print("Child Class: Buying Smartphone")

s = SmartPhone('Android', 12)
s.buy() #Printed the child class method. 
````

## Super Keyword 
- In case where Parent and child class both of them has the constructor and we need to initilize both the constructor using inheritence. Then we use the super() keyword, allows us to call the constructor of the parents class also. 

### Note:
1. super() cannot access variable
2. super() written inside the child class
3. super() cannot be written outside the class. 

````python
class Phone:
    def __init__(self, price, brand, camera):
        print("Parents class: inside Phone constructor")
        self.price = price 
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Parents Class: buying Phone")

class SmartPhone(Phone):
    def __init__(self,price, brand, camera, os, ram):
        print("Child class: Inside Smartphone constructor")
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram

    def buy1(self):
        print("Child Class: Buying smartphone")

s = SmartPhone(2000, 'Realme', 12, 'Android', 12)
s.buy() # both of them will print the child class method only, due to method overriding
s.buy1()
````

## Types of inheritance
- There are in total 5 types of inheritance in python 
1. Single Inheritance
2. Multi level Inheritance
3. Hierarchical Inheritance
4. Multiple Inheritance
5. Hybrid Inheritance

### Single Inheritance 

``` mermaid
    A[Parent]----->B[Child];
```
- Only inheriting from its parent

````python
# Sigle Inheritance 

class A: 
    def __init__(self):
        print('Parent Class: Inside constructor')
        self.a = 'Realme'

    def __example_method(self):
        print('Inheriting from parents class')

    def get_example_method(self):
        return self.__example_method()

class B(A):
    def __init__(self):
        super().__init__()
        print('child Class: Inside constructor')
        self.b = 'Child Realme' 

a = B()
a.get_example_method() # Inheriting the parents class method. 
````

### Multilevel Inheritance
- Inheritance is being performed in multiple level
- Chan can be infinitely long in any direction
- Child can inherit the members of all its parents at the same time

``` mermaid
    .... A[Parent]<------B[Parent];
    B[Parent]<------C[Child] ...;
```

````python
# Multi level Inheritance 

class A: 
    def __init__(self):
        print('Parent2 Class: Inside constructor')
        self.a = 'Realme'

    def __example_method(self):
        print('Grand Parent class method is accessible by the child class.')

    def get_example_method(self):
        return self.__example_method()

class B(A):
    def __init__(self):
        super().__init__()
        print('Parent1 Class: Inside constructor')
        self.b = 'Child Realme' 

    def __parent_method(self):
        print("Parent 1 Class method is also accessible by the child")

    def get_parent_method(self):
        return self.__parent_method()

class C(B):
    print("Child Class: Has nothing inside this class")
    pass

a = C()
# a.get_example_method() # Inheriting the parents class method.

#Two methods grand parents method and parents class method, which is accessible by the child classs
a.get_example_method() #Grand Parent class method
a.get_parent_method() #Parent class method
````

### Hierarchical Inheritance
- Multiple childern can inherit properties from a single parent
- Any number of child can inherit from the parents class. 

``` mermaid
    A[Parent-1]<------B[child-1];
    A[Parent-1]<------C[Child-2];
    A[Parent-1]<------D[Child-3];
```

````python
# Hierarchical Inheritance 

class A: 
    def __init__(self):
        print('Parent1 Class: Inside constructor')
        self.a = 'Realme'

    def __example_method(self):
        print('Parent-1 class method is accessible by the child-1 class as well as child-2.')

    def get_example_method(self):
        return self.__example_method()

class B(A):
    def __init__(self):
        super().__init__()
        print('Child-1 Class: Inside constructor')
        self.b = 'Child Realme'

class C(A):
    def __init__(self):
        super().__init__()
        print('Child-2 Class: Inside constructor')
        self.b = 'Child Realme' 

a = B() #child object 1
b = C() #child object 2

#Two methods grand parents method and parents class method, which is accessible by the child classs
a.get_example_method() #Grand Parent class method
b.get_example_method() #Parent class method
````

### Multiple Inheritance 
- Single child inheriting from multiple parents.

__Note__: (Method Resolution order)
- When child class is inheriting multiple class, it will check the order of the class pass inside the child. Based on that it will decide which classes method to execute first.

``` mermaid
    A[Parent-1]<------D[child-1];
    B[Parent-2]<------D[Child-1];
    C[Parent-3]<------D[Child-1];
```

### Hybrid Inheritance
- In this type of inheritance we are using __Multi level Inheirtance, Hierarchical Inheritance, and Multiple Inheritance and Single Inheritance__ hence it is called hybrid inheritance.

````python
# Multiple Inheritance 

class A: 
    def __init__(self):
        print('Parent-1 Class: Inside constructor')
        # super().__init__()
        self.a = 'Realme'

    def __example_method(self):
        print('Parent-1 class method is accessible by the child class.')

    def get_example_method(self):
        return self.__example_method()

class B:
    def __init__(self):
        # super().__init__()
        print('Parent-2 Class: Inside constructor')
        self.b = 'Child Realme' 

    def __parent_method(self):
        print("Parent-2 Class method is accessible by the child class")

    def get_parent_method(self):
        return self.__parent_method()

# This is called method resolution order 
class C(A, B):
    pass

class D(B, A):
    pass 

print("\n called by 'C' class object")
a = C()
a.get_example_method() # Parent class method
a.get_parent_method() # Parent class method
print("\n called by 'D' class object")
b = D()
b.get_example_method()
b.get_parent_method()
````
