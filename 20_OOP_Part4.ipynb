# Polymorphism 

- It means that, depending on the situation, different methods, operators behaves differently.

## It majorly have 3 concepts.
1. Method Overriding
2. Method Overloading
3. Operator Overloading

### Method Overriding 
- Two class having same method with the same class name. By default it will call the childs class method.

### Method Overloading
- In a single class if two methods having same name, depending on the input it will call the methods. 
- Unfortunately, Python does not support Method Overloading, still we can implement in several ways. 

### Operator Overloading
- If same operator is used for different purpose and gives different output, then it is called operator overloading. 

```python
#This is called method overriding
class Phone: 
    def __init__(self, price, brand, camera):
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("parent Class: Buying Phone")

class SmartPhone(Phone):
    def __init__(self, os, ram):
        self.os = os
        self.ram = ram

    def buy(self):
        print("Child Class: Buying Smartphone")

s = SmartPhone('Android', 12)
s.buy() #Printed the child class method. 
```

```python
# This is method overloading
class A: 
    def __init__(self):
        print("only one class")
    
    # Here we have initialised 'b' with 0, this is known as default method.
    def area(self, a, b=0):
        if b == 0: return 3.14*(a)**2 
        else: return a*b 


obj = A()
print(obj.area(2))
print(obj.area(2, 4))            
        
```

```python
# This is called operator overloading. 
class Addition: 
    def __init__(self, a):
        self.a = a 

    def __add__(self, other):
        return Addition(self.a + other.a)

    def __str__(self):
        return str(self.a)

a = Addition('Pratyush')
b = Addition('Saha')
print(a + b)

c = Addition(5)
d = Addition(6)
print(c + d)

e = Addition([1, 2, 3, 4, 5])
f = Addition([6, 7, 8, 9, 10])
print(e + f)

print("\n-> using __str__")
print(a.__str__())
print(b.__str__())
print(c.__str__())
print(d.__str__())
print(e.__str__())
print(f.__str__())
```
