# Polymorphism 

**Polymorphism** comes from two Greek words: *Poly* (meaning "many") and *Morphism* (meaning "forms").
In programming, it simply means having **one name, but many forms or behaviors**.

Imagine a "speak" command:
- If you give it to a dog, the dog barks.
- If you give it to a cat, the cat meows.
- If you give it to a duck, the duck quacks.
The command ("speak") is exactly the same, but the behavior changes depending on the object!

## The 3 Main Concepts of Polymorphism
There are three common ways we see polymorphism in action:
1. **Method Overriding**
2. **Method Overloading**
3. **Operator Overloading**

---

### 1. Method Overriding (Replacing a Parent's Behavior)
- **What it is:** When a child class has a method with the *exact same name* as a method in its parent class.
- **Why we use it:** The child class doesn't want to use the parent's default behavior, so it creates its own custom version to "override" it. By default, the child's version will be called.

### 2. Method Overloading (Handling Different Inputs)
- **What it is:** Having multiple methods with the *same name* in a single class, but they take different types or numbers of inputs (arguments). The program knows which one to call based on the input you give.
- **Python Note:** *Python does not support true method overloading by default.* However, we can use neat tricks (like default arguments) to achieve the same result!

### 3. Operator Overloading (Giving Symbols New Jobs)
- **What it is:** Giving a standard mathematical operator (like `+`, `-`, or `*`) a new meaning depending on what you are working with.
- **Example:**
  - If you use `+` with numbers (`5 + 5`), it performs **addition** (output: `10`).
  - If you use `+` with strings (`"Hello " + "World"`), it performs **joining/concatenation** (output: `"Hello World"`).
  
  The same `+` symbol is acting differently depending on the situation!

```python
#This is called method overriding
class Phone: 
    def __init__(self, price, brand, camera):
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("parent Class: Buying Phone")

class SmartPhone(Phone):
    def __init__(self, os, ram):
        self.os = os
        self.ram = ram

    def buy(self):
        print("Child Class: Buying Smartphone")

s = SmartPhone('Android', 12)
s.buy() #Printed the child class method. 
```

```python
# This is method overloading
class A: 
    def __init__(self):
        print("only one class")
    
    # Here we have initialised 'b' with 0, this is known as default method.
    def area(self, a, b=0):
        if b == 0: return 3.14*(a)**2 
        else: return a*b 


obj = A()
print(obj.area(2))
print(obj.area(2, 4))            
        
```

```python
# This is called operator overloading. 
class Addition: 
    def __init__(self, a):
        self.a = a 

    def __add__(self, other):
        return Addition(self.a + other.a)

    def __str__(self):
        return str(self.a)

a = Addition('Pratyush')
b = Addition('Saha')
print(a + b)

c = Addition(5)
d = Addition(6)
print(c + d)

e = Addition([1, 2, 3, 4, 5])
f = Addition([6, 7, 8, 9, 10])
print(e + f)

print("\n-> using __str__")
print(a.__str__())
print(b.__str__())
print(c.__str__())
print(d.__str__())
print(e.__str__())
print(f.__str__())
```
