# File Handling
--- 
### Types of __Data__ used for I/O (Input/Output):
1. **Text:** Human-readable text (e.g., `'12345'`, `'Hello World'`). These are stored as a sequence of **Unicode** characters.
2. **Binary:** Computer-readable data (e.g., `12345` as raw memory bytes like `0x89`). These are stored as a sequence of **bytes**.

### Types of **Files** we deal with:
- **Text Files:** Files you can open and read in a standard text editor (e.g., `.txt`, `.py`, `.csv`). They use encodings like `utf-8`.
- **Binary Files:** Media and executable files (e.g., `.jpg` (Images), `.mp4` (Videos), `.exe` (Programs)). You need specific software to open these!

--- 

### The 3 Steps of File I/O
Think of file handling like reading a physical book:
1. **Step 1: Open the file** *(Taking the book off the shelf and opening it)*
2. **Step 2: Read/Write data** *(Reading the story or writing notes in the margins)*
3. **Step 3: Close the file** *(Closing the book and putting it back on the shelf)*

# Working with Text Files

### How does the `open()` function work?

1. **Opening a file:** When we run `open('dummy.txt', 'w')`, Python fetches the file from your **Storage** (Hard Drive/SSD) and loads it into **RAM** (active memory) inside a temporary holding area called a **Buffer**.
2. **Closing a file:** When we run `close()`, Python takes any updated data from the RAM buffer and permanently saves it back to your Storage. It then safely frees up the memory.

> **What is a Buffer?** 
> A buffer is a temporary memory area. Instead of interacting with the slow hard drive one single character at a time, Python groups the data into "chunks" in the buffer and processes them all at once. This makes file handling much faster!

---

### The Danger of the `'w'` (Write) Mode
If you open an existing file in `'w'` mode, **it will completely erase the old data** and overwrite it with your new content! 

If you want to keep your existing data and just add new text to the very end of the file, you should use the **`'a'` (Append) mode** instead.

```python
# 'a' mode opens the file without deleting the existing content
f = open("demo.txt", "a") 

# This text will be added to the bottom of the file
f.write("\nHi, I am Pratyush")

f.close()


````python
L = ["\nI am a software Engineer\n", "I love coding\n", "I love to solve new challenges"]

f = open("22_file_handling_output/demo.txt", "a")
f.writelines(L)
f.close()
````

### Reading from a File

There are two primary ways to extract data from a text file:
1. `read()` : Reads the **entire file** all at once.
2. `readline()` : Reads the file **one line at a time**.

> **Remember:** When you read from a text file, Python will always return the data to you as a **String**, even if the file contains numbers! If you need to do math with those numbers, you will have to convert them (e.g., `int(data)`).

**1. Using the `read()` function:**
The `read()` function grabs everything inside the file and stores it into a single string variable.

```python
# 1. Open the file in 'r' (read) mode
f = open("demo.txt", "r")

# 2. Read the ENTIRE file and store it in variable 's'
s = f.read()

# 3. Print the contents to the console
print(s)

# 4. Always close the file!
f.close()


````python
f = open("22_file_handling_output/demo.txt", "r")
s = f.read()
print(s)
f.close()
````

````python
# Reading only few or upto 'n' characteristics
f = open("22_file_handling_output/demo.txt", 'r')
s = f.read(15)
f.close()
print(s)
````

**2. Using the `readline()` function:**
If you don't want to load a massive file into memory all at once, you can read it one line at a time using `readline()`.
```python
f = open("22_file_handling_output/demo.txt", 'r')
# Read and print the first two lines
print(f.readline(), end='')
print(f.readline(), end='')
f.close()
```
>**The Double Newline Mystry: Why use end=''**? <br/>

- When you use readline(), it grabs the line of text along with the hidden newline character (\n) at the end of it.<br/>
- Because the print() function also automatically adds its own newline by default, printing a line from a file normally results in two newlines (creating an ugly blank gap between your lines).<br/>
- By adding end='' inside the print() function, we are telling Python: "Hey, don't add your own newline this time!" This prevents the extra gap in your output.

````python
# Reading line by line using 'Readline()' function
print("In this you'll see line gap in the output:")
f = open("22_file_handling_output/demo.txt", 'r')
print(f.readline())
print(f.readline()) 
f.close()

print("In this you'll not see line gap in the output:")
f = open("22_file_handling_output/demo.txt", 'r')
print(f.readline(), end='')
print(f.readline(), end='') 
f.close()
````

### Reading an Entire File Line-by-Line (Using Loops)

If a file has hundreds of lines, we obviously can't type `f.readline()` manually for every single line! 

Instead, we can combine `readline()` with a **`while` loop** to automatically read through the file until there is no data left.


````python
# Printing message from a file, using 'readline()' function
f = open("22_file_handling_output/demo.txt", 'r')

while True:
    data = f.readline()
    if data == '':
        break
    else:
        print(data, end='')
f.close()
````

# The Smarter Way: Context Managers (`with`)

Up until now, we've had to manually type `f.close()` every time we finished working with a file. But what if we forget? *(Spoiler: It happens to the best of us!)*

To prevent files from being accidentally left open, Python gives us a smarter way to handle files using the **`with`** keyword, which is technically called a **Context Manager**.

### Why use `with`?
- **Automatic Closing:** It automatically closes the file for you as soon as the indented block of code is finished running! You never have to write `f.close()` again.
- **Cleaner Code:** It is widely considered the "Pythonic" (best practice) way to write file handling code in the industry.


````python
# using context manager "with", writing using "write",

with open("22_file_handling_output/demo1.txt", 'w') as file:
    file.write("Hey man!")

# reding a file with contex manager. 
with open("22_file_handling_output/demo1.txt", 'r') as file:
    text = file.read()
    print(text)
````

### Reading Specific Chunks (The File Cursor)

When you use `read(n)`, you are telling Python to read exactly `n` characters from the file. But how does Python know where to pick up on the next read?

Think of Python as having an invisible **Cursor** (or bookmark) inside the file:
- **First Read:** `file.read(10)` reads the first 10 characters. The invisible cursor then stops and waits at character 10.
- **Second Read:** If you call `file.read(5)` immediately after, it doesn't start from the beginning! It picks up exactly where the cursor left off and reads the *next* 5 characters.


````python
# Moving in a file, character by character.

smallFile = ["\nI am a software Engineer\n", "I love coding\n", "I love to solve new challenges"]

with open("22_file_handling_output/demo1.txt", 'w') as file:
    file.writelines(smallFile)

with open("22_file_handling_output/demo1.txt", 'r') as file:
    text = file.read(20)
    text2 = file.read(20)
    print(text)
    print(text2)
````

````python
# Writing a big file with context manager. 
bigFile = ["Hi, I am Pratyush \n" for sentence in range(1000)]

with open("22_file_handling_output/demo.txt", "w") as file:
    file.writelines(bigFile)
````

## Controlling the Cursor: `seek()` and `tell()`

Now that we know Python uses an invisible "Cursor" to keep track of where it is in a file, what if we want to manually move that cursor around?

- **`tell()` : "Where am I?"** 
  This function tells you the exact current position (index) of the cursor inside the file.
  
- **`seek(n)` : "Go to position `n`!"**
  This function teleports the cursor to a specific character position `n`. The next time you read or write, it will start from this new location.

> **NOTE:** Using `seek()` **does not erase or remove** any characters in the file! It is purely for moving the cursor around, exactly like clicking your mouse in the middle of a Word document.


````python
# Reading with seek() and tell():
with open("22_file_handling_output/demo.txt", "r") as file:
    print(file.tell()) # At the starting the pointer is at 0th position.
    file.seek(200)
    print(file.read(50))
    print(file.tell()) # this tells at which position the pointer is pointing. 
````

````python
# Writing with seek() and tell():
smallFile = ["\nI am a software Engineer\n", "I love coding\n", "I love to solve new challenges"]

with open("22_file_handling_output/demo1.txt", "w") as f: 
    f.writelines(smallFile)
    print(f.seek(30))
    print(f.write(" Seek() ")) # write also prints the number of character that is being written in the text file
    print(f.tell())
     
````

## File I/O with Error Handling (`try-except`)

### Case 1: Reading a File Safely
Here, we try to open and read a file. If the file doesn't exist, Python will instantly jump to the `except` block and print our custom message instead of crashing with a massive red error.

```python

    print(file.readlines())
except:
    # This runs if any error occurs (like the file not existing)

else:
    # This runs ONLY if the 'try' block succeeds without any errors
    file.close()
    print("file closed")
```

# Working with Binary Files

So far, we have only worked with text files (strings). But what if you want to write a program that interacts with an image, a video, or an audio file? These are called **Binary Files**.

### Reading and Writing in Byte Mode
Because binary files are not made of human-readable text, we cannot use strings. We must read and write them using raw **bytes**. To do this, we add the `b` (binary) flag to our modes:
- `'rb'` : Read Binary
- `'wb'` : Write Binary

### Example: Copying an Image
The most common way to practice binary file handling is to write a script that creates a duplicate of an image. 

To do this, we simply:
1. Open the original image in `'rb'` (Read Binary) mode.
2. Create a brand new, empty file in `'wb'` (Write Binary) mode.
3. Read the bytes from the original file, and write those bytes directly into the new copy!


````python
with open("22_file_handling_output/Screenshot.png", "rb") as file:
    with open("22_file_handling_output/screenshot_copy.png", "wb") as writefile:
        writefile.write(file.read())
````

# Serialization & Deserialization

As we saw earlier, the **major issue** with standard text files is that they **only understand strings**. We couldn't easily save a complex Python Dictionary or a List without breaking it.

To solve this problem, we use a computer science concept called **Serialization and Deserialization**.

### 1. Serialization (Writing / "Dumping")
- **What it is:** The process of converting complex Python data types (like dictionaries or lists) into a structured text format so they can be safely saved to a file.
- **Our Tool:** We will use the **JSON** format to serialize our data. In Python, we call this "dumping" the data.

### 2. Deserialization (Reading / "Loading")
- **What it is:** The exact opposite! It is the process of reading that structured text from a file and converting it *back* into a usable Python data type (like a working dictionary).
- **Our Tool:** We read the JSON file and "load" it back into our Python code. 

> **Why JSON?** JSON (JavaScript Object Notation) is the most popular data format in the programming world. If you serialize your Python data into a JSON file, a website built in JavaScript or an Android app built in Java can easily read and use your data!


````python
# Serialization (with List data type) 
import json 
import os 

List = [1,2,3,4,5,6]

with open("demo_json.json", "w") as file:
    json.dump(List, file) 

if os.path.exists("demo_json.json"):
    os.remove("demo_json.json") #removed the file. 
    print("File removed successfully!")

with open("22_file_handling_output/demo_json.json", "w") as file: # This dumping of data is called serialization. 
    json.dump(List, file)
    print("New demo_json file created inside the file_handling_output folder")
````

````python
# Deserialization ( With List data type )
import json 

with open("22_file_handling_output/demo_json.json", "r") as file:
    read = json.load(file) 
    print(read)
````

````python
# Serialization (with dictionary data type)
import json 

dictionary = {
    "name": "pratyush", 
    "age": "22", 
    "gender": "male" 
}

with open("22_file_handling_output/demo_json.json", "w") as file: 
    json.dump(dictionary, file, indent=4) 
````

## Serialization & Deserialization of Custom Objects 

So far, we have serialized basic Python data types like lists and dictionaries. But what if you create your own custom Class (like a `Student` or a `Car`)? 

Can we save an entire custom Object (along with all of its attributes) into a file, and then load it back into our code later? Let's find out!


````python
# Let us do with Dictionary Format first. 
class Car:
    def __init__(self, model, color, brand, price):
        self.model = model 
        self.color = color
        self.brand = brand
        self.__price = price

    def get_price(self):
        return self.__price

car = Car('M340i', 'Grey', 'BMW', '7400000')

def show_obj(car):
    if isinstance(car, Car):
        return {'Mode':car.model, 'color':car.color, 'Brand':car.brand, 'Price':car.get_price()}


with open('22_file_handling_output/demo_json1.json', 'w') as file:
    json.dump(car, file, default = show_obj, indent=4)


with open('22_file_handling_output/demo_json1.json', 'r') as file: 
    car = json.load(file)
    print(car)  
    print(type(car))  

````

````python
def show_obj1(car):
    if isinstance(car, Car):
        return (car.model, car.color, car.brand, car.get_price())
    else:
        print("Cannot write")


with open('22_file_handling_output/demo_json1.json', 'w') as file:
    json.dump(car, file, default = show_obj1, indent=4)


with open('22_file_handling_output/demo_json1.json', 'r') as file: 
    car = json.load(file)
    print(car)  
    print(type(car))  
````

# Pickling (Python's Magic Trick)

While JSON is amazing for basic data, it struggles to save **Custom Objects** (like our custom classes). To save complex Python objects, we use a built-in Python library called **Pickle**.

- **Pickling:** The process of converting any Python object into a **byte stream** (binary format) so it can be saved to a file. 
- **Unpickling:** Loading that byte stream back into a working Python object!

> **Important:** Because Pickling converts objects into binary (bytes), you must always use the `'wb'` (write binary) and `'rb'` (read binary) modes when opening your files!

> **Pro Tip:** Unlike JSON, Pickling is specific to Python. If you save a file using Pickle, only another Python program will be able to read it. You cannot send a Pickled file to a JavaScript or Java program!


````python
import pickle
import pickle

class Car:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.__price = price

    def get_price(self):
        return self.__price

    def display(self):
        print(f"My Car is {self.brand} {self.model}, it costs around {self.get_price()}!")
        return

car = Car('BMW', 'M340i', 7400000)

with open('22_file_handling_output/car.pkl', 'wb') as file: 
    pickle.dump(car, file)
    
````

# Unpickling 

Once we have successfully saved our custom object as a binary file, how do we get it back into our code? We Unpickle it!

- **Unpickling:** The process of reading a byte stream (a binary file) and converting it back into a fully functional Python object. 
- **The Code:** We use the `pickle.load()` function to achieve this.

> **Reminder:** Since we are reading from a binary file, make sure to open the file using the `'rb'` (Read Binary) mode!


````python

with open("22_file_handling_output/car.pkl", "rb") as file:
    carData = pickle.load(file) 
    carData.display()
````

### Pickle vs JSON

Here is a quick cheat sheet summarizing the differences between the two serialization methods we just learned:

| Feature | Pickle | JSON |
| :--- | :--- | :--- |
| **Format** | Stores data in a **Binary** (byte) format. | Stores data in a **Human-readable** text format. |
| **Compatibility** | Only works with **Python** programs. | Works with **almost all programming languages** (Universal). |
| **Best Used For** | Saving complex **Custom Objects** (like custom Classes). | Saving standard data types (like Dictionaries and Lists). |
| **Security** | Can be unsafe if unpickling files from an untrusted source. | Generally very safe to read and load. |
