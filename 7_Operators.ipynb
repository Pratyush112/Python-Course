<h1>Operators</h1>
<ol>
    <li><strong>Arithmetic Operators:</strong> These operators are used for performing arithmetic operations on numbers.
        <ul>
            <li>Addition(+),</li>
            <li>Subtraction(-)</li>
            <li>Multiplication(*)</li>
            <li>Division(/)</li>
            <li>Power of (**)</li>
            <li>Modulus(%)</li>
            <li>Integer Division(//)</li>
        </ul>
    </li>
    <li><strong>Comparison Operators:</strong> These operators are used for comparing values.
        <ul>
            <li>EQUAL TO(==),</li>
            <li>NOT EQUAL TO(!=)</li>
            <li>Greater Than(>)</li>
            <li>Less Than(<)</li>
            <li>Greater than or equal to (>=)</li>
            <li>Less that or equal to (<=)</li>
        </ul>
    </li>
    <li><strong>Logical Operator:</strong> These operators are used for combining logical expressions.
        <ul>
            <li>or</li>
            <li>and</li>
            <li>not</li>
        </ul>
    </li>
    <li><strong>Bitwise Operators:</strong> perform operations directly on the individual bits (binary representations) of integers.
        <ul>
            <li>'&' -> and</li>   
            <p>Example: 5 &amp; 3 = 1 (101 & 011 = 001)</p>
            <li>'|' -> or</li>
            <p>Example: 5 | 3 = 7 (101 & 011 = 111)</p>
            <li>'>>' -> Right shift</li>
            <p>Example: 5 >> 3 = 0 (101 & 011 = 0000) => 5 // 2^3 => 5//8 = 0</p>
            <li>'<<' -> Left shift</li>
            <p>Example: 5 << 3 = 40 (101 & 011 = 0111) => 5 * 2^3 => 5*8 = 40</p>
            <li>'~' -> NOT</li>
            <p>Example: ~5 = -6 (~000000101 = 11111010)<br> => Python uses signed integers (two’s complement representation), so it can represent negative numbers as well. <br> 
            <strong>The Rule:</strong> x = -(x + 1) <br> Therefore,<br> x = 5, <br> 5 = -(5+1) = -(6) = -6</p>
            <li>'^' -> XOR (exclusive or)</li>
            <p>Example: 5 ^ 3 = 6 (0101 & 0011 = 0110)</p>
        </ul>
    </li>
    <li><strong>Assignment Operators:</strong> These operators are used for assigning values to variables.
        <ul>
            <li>=</li>
        </ul>
    </li>
    <li><strong>Identity Operators:</strong> Check whether the two variables are in the same memory location.
        <ul>
            <li>is - Returns True if both variables are the same</li>
            <li>is not - Returns False if both variables are the same</li>
        </ul>
    </li>
    <li><strong>Membership Operators:</strong> used to test whether a value or variable is found within a sequence such as a string, list, tuple, set, or dictionary
        <ul>
            <li>is - Returns True if both variables are the same</li>
            <li>is not - Returns False if both variables are the same</li>
        </ul>
    </li>
</ol>

```python
print("Arithmetic Operators:")
print("Addition: 5 + 3 =", 5 + 3)  # Output: 8
print("Subtraction: 5 - 3 =", 5 - 3)  # Output: 2
print("Multiplication: 5 * 3 =", 5 * 3)  # Output: 15
print("Division: 5 / 3 =", 5 / 3)  # Output: 1.6666666666666667
print("Floor Division: 5 // 3 =", 5 // 3)  # Output: 1
print("Modulus: 5 % 3 =", 5 % 3)  # Output: 2
print("Exponentiation: 5 ** 3 =", 5 ** 3)  # Output: 125

print("\nComparison Operators:")
print("Equal to: 5 == 3 ->", 5 == 3)  # Output: False
print("Not equal to: 5 != 3 ->", 5 != 3)  # Output: True
print("Greater than: 5 > 3 ->", 5 > 3)  # Output: True
print("Less than: 5 < 3 ->", 5 < 3)  # Output: False    
print("Greater than or equal to: 5 >= 3 ->", 5 >= 3)  # Output: True
print("Less than or equal to: 5 <= 3 ->", 5 <= 3)  # Output: False

print("\nLogical Operators:")
print("Logical OR: (5>3) OR (5<3):", (5>3) or (5<3))  # Output: True
print("Logical AND: (5 > 3 and 5 < 3)", 5 > 3 and 5 < 3)  # Output: False
print("Logical NOT: (not (5 > 3))", not (5 > 3))  # Output: False

print("\nBitwise Operators:")
a = bin(5) 
print("Binary of 5: ", a)
b = bin(3)
print("Binary of 3:", b)
print("Bitwise AND: 5 (Binary - 101) & 3 (Binary - 011) = ", 5 & 3)  # Output: 1
print("Bitwise OR: 5 | 3 = ", 5 | 3)   # Output: 7
print("Bitwise XOR: 5 ^ 3 = ", 5 ^ 3)  # Output: 6
print("Bitwise NOT: ~5 = ", ~5)        # Output: -6
print("Left Shift: 5 << 5 = ", 5 << 5) # Output: 160
print("Right Shift: 5 >> 5 = ", 5 >> 5) # Output: 0
```
